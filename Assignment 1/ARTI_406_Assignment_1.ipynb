```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
path = kagglehub.dataset_download("rounakbanik/pokemon") #this is a dataset of pokemon cards
print("Path:", path)
```

```python
df = pd.read_csv(os.path.join(path, "pokemon.csv")) #use os.path.join for so its more safer then absolute path.
df.head(5) #reads the first 5
```

```python
df.info() #prints the info of the dataframe
```

```python
df.duplicated()[df.duplicated()==True] # Means there is no duplicates
```

```python
#Checking missing values
print("Missing values:")
print(df.isnull().sum())
```

```python
df.describe(include='all') #stat summary
```

# Distributions

```python
#Distributions
plt.figure(figsize=(10, 5))
plt.hist(df['attack'].dropna(), bins=30, edgecolor='black')
plt.title('Attack Distribution')
plt.xlabel('Attack')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df['sp_attack'].dropna(), bins=30, edgecolor='black', color='yellow')
plt.title('SP Attack Distribution')
plt.xlabel('sp_attack')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df['speed'].dropna(), bins=30, edgecolor='black', color='orange')
plt.title('Speed Distribution')
plt.xlabel('speed')
plt.ylabel('Frequency')
plt.show()
```

# Removing NaN

# Attack vs Defense Relationship

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='attack', y='defense', data=df)
plt.title("Attack vs Defense")
plt.show()
```
